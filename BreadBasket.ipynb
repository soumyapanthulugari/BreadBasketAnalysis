```python
import pandas as pd
from mlxtend.frequent_patterns import apriori 
from mlxtend.frequent_patterns import association_rules
```

```python
bread=pd.read_csv("/content/bread basket.csv")
```

```python
bread.head()
```

```python
bread.shape
```

```python
bread['weekday_weekend'].value_counts()
```

```python
bread['Item']=bread['Item'].str.strip()
bread.dropna(axis=0,subset=['Transaction'],inplace=True)
bread['Transaction']=bread['Transaction'].astype('str')
bread.shape
```

```python
bread['weekday_weekend'].value_counts()
```

```python
bread[['period_day','weekday_weekend']].value_counts()
```

#Weekday_weekend

```python
mybaskets= bread.groupby('Transaction')['Item'].value_counts().unstack().reset_index().fillna(0).set_index('Transaction')
mybaskets
```

##converting all positive vaues to 1 and everything else to 0

```python

def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

weekend_weekday = mybaskets.applymap(my_encode_units)
```

##Generating frequent itemsets

```python

freq_items= apriori(weekend_weekday, min_support=0.004, use_colnames=True)
```

##generating rules

```python

weekday_weekend_rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
```

```python
weekday_weekend_rules
```

##Making recommendations for Weekday_weekend 

```python
mybaskets['Keeping It Local'].sum()
```
